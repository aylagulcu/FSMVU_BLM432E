```python
data_file = open('strb.csv',encoding="utf8")
lats, lons = [], []
types = []
for index, line in enumerate(data_file.readlines()):
    if index > 1:
        try:
            types.append(float(line.split(',')[1]))
            lons.append(float(line.split(',')[2]))
            lats.append(float(line.split(',')[3]))
        except ValueError:
            pass
def get_marker_color(x):
    if x == 1.0:
        return ('go')
    elif x== 2.0:
        return ('yo')
    else:
        return ('ro')

w_map = Basemap(projection='robin', resolution = 'l',
              lat_0=0, lon_0=0)

 
w_map.drawcoastlines()
w_map.drawcountries()
w_map.fillcontinents()
w_map.drawmapboundary()
w_map.drawmeridians(np.arange(0, 360, 30))
w_map.drawparallels(np.arange(-90, 90, 30))


for lon, lat, ty in zip(lons, lats, types):
    x,y = w_map(lon, lat)
    marker_string = get_marker_color(ty)
    w_map.plot(x, y, marker_string)

plt.show()
```
