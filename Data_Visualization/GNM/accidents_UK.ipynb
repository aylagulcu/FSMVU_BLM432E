```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
import folium
import seaborn as sns

#sns.set(style="white", color_codes=True)
```

```python
accidents= pd.read_csv("DfTRoadSafety_Accidents_2014.csv")
```

```python
fig, ax= plt.subplots(figsize=(8,8))
m = Basemap(resolution='f',
            projection='merc',
            lat_0=54.5, lon_0=-3,
            llcrnrlon=-6.5, llcrnrlat= 49.5, urcrnrlon=2., urcrnrlat=59)
             
m.drawmapboundary(fill_color='#46bcec')
m.fillcontinents(color='#f2f2f2',lake_color='#46bcec')
m.drawcoastlines()
m.readshapefile('UK-postcode-boundaries-Jan-2015/Distribution/Areas', 'areas')
x, y = m(list(accidents["Longitude"].astype(float)), list(accidents["Latitude"].astype(float)))
m.plot(x, y, 'o',markersize=2 , alpha=0.5)
```
